```python

df['Age'] = df['Age'].str.replace(' years old', '', regex=False)

df['Age'] = df['Age'].replace('65 years or older', '65+')

age_mode = df['Age'].mode()[0]
df['Age'] = df['Age'].replace('Prefer not to say', age_mode)


print(df['Age'].value_counts())
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

df['Country'] = df['Country'].replace('United States of America', 'USA')
df['Country'] = df['Country'].str.replace('United Kingdom of Great Britain', 'UK')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Country', order=df['Country'].value_counts().index[:10], palette="Set2")
plt.title("Top 10 Countries of Respondents")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()

```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
print(df['EdLevel'].unique())
```

```python

df['EdLevel'] = df['EdLevel'].str.replace(' (e.g. American high school, German Realschule or Gymnasium, etc.)', '', regex=False)
df['EdLevel'] = df['EdLevel'].str.replace(' (B.A., B.S., B.Eng., etc.)', '', regex=False)
df['EdLevel'] = df['EdLevel'].str.replace(' (M.A., M.S., M.Eng., MBA, etc.)', '', regex=False)
df['EdLevel'] = df['EdLevel'].str.replace('Some college/university study without earning a degree', 'Attended college but no degree', regex=False)
df['EdLevel'] = df['EdLevel'].str.replace(' (JD, MD, Ph.D, Ed.D, etc.)', '', regex=False)
df['EdLevel'] = df['EdLevel'].str.replace(' (A.A., A.S., etc.)', '', regex=False)
df = df[df['EdLevel'] != 'Something else']

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='EdLevel', order=df['EdLevel'].value_counts().index, palette="Set2")
plt.title("Distribution of Education Levels")
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.show()
```

```python
df['YearsCode'] = df['YearsCode'].replace('Less than 1 year', '<1')
df['YearsCode'] = df['YearsCode'].replace('More than 50 years', '50+')
plt.figure(figsize=(22, 6))

# Use a barplot for categorical counts
sns.barplot(x=df['YearsCode'].value_counts().index, y=df['YearsCode'].value_counts().values, color='green', width=0.7)

# Rotate x-axis labels for better readability
#plt.xticks(rotation=45)

plt.title("Years of Coding Experience Distribution")
plt.xlabel("Years of Experience")
plt.ylabel("Frequency")
plt.show()
```

```python

language_counts = df['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.values[:10], y=language_counts.index[:10], palette="muted")
plt.title("Top 10 Programming Languages Used")
plt.xlabel("Count")
plt.ylabel("Language")
plt.show()
```

```python
df = df[df['Industry'] != 'Other:']
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Industry', order=df['Industry'].value_counts().index[:10], palette="Set2")
plt.title("Top 10 Industries of Respondents")
plt.xlabel("Count")
plt.ylabel("Industry")
plt.show()
```

```python
df['OrgSize'] = df['OrgSize'].replace('Just me - I am a freelancer, sole proprietor, etc.', 'Self-employed')
df = df[df['OrgSize'] != 'I don’t know']
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='OrgSize', order=df['OrgSize'].value_counts().index, palette="coolwarm")
plt.title("Organization Size Distribution")
plt.xlabel("Count")
plt.ylabel("Organization Size")
plt.show()
```

```python

plt.figure(figsize=(22, 6))
sns.scatterplot(data=df, x='YearsCode', y='Age', alpha=0.7, color='purple')
plt.title("Age vs. Years of Coding Experience")
plt.xlabel("Years of Coding Experience")
plt.ylabel("Age")
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='EdLevel', y='Age', palette="Set3")
plt.title("Age Distribution Across Education Levels")
plt.xlabel("Education Level")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.show()
```
